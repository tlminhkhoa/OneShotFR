```python
import numpy as np
from sklearn.svm import SVC
from PIL import Image
import numpy as np
from mtcnn.mtcnn import MTCNN
from keras.models import load_model
from numpy import savetxt, loadtxt
import pandas as pd
import os
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
def format_for_facenet(imface):
    imface = imface.resize((160, 160))
    face_pixels = np.asarray(imface)
    mean, std = face_pixels.mean(), face_pixels.std()
    face_pixels = (face_pixels - mean) / std
    face_pixels = np.expand_dims(face_pixels, axis=0)
    return face_pixels
```

```python
def EuclideanDistantClassifier(test, threhold):
    min = 100
    identity = None
    dist = []
    for i in range(0, len(trainset)):
        current = np.linalg.norm(trainset.iloc[i, :-1]-(test))
        dist.append(current)
        if (current < min):
            min = current
            identity = trainset.iloc[i, -1]

    print(dist)
    if (min > threhold):
        return "unknown"
    else:
        return identity
```

```python
def load_train_data(directory):
    df = pd.DataFrame()
    labels = []
    for filename in os.listdir(directory):
        if filename.endswith(".csv"):
            if(df.empty):
                df = pd.DataFrame(loadtxt(os.path.join(
                    directory, filename), delimiter=',').reshape(1, 128))
            else:
                tempdf = pd.DataFrame(loadtxt(os.path.join(
                    directory, filename), delimiter=',').reshape(1, 128))
                df = pd.concat([df, tempdf])
            labels.append(filename.replace(".csv", ""))

    df["labels"] = labels

    return df
```

```python
def extract_face(impixel, boxReturn):
    box_det = facedetector.detect_faces(impixel)
    if box_det:
        box = box_det[0].get('box')

        if (boxReturn == True):
            return box

        x1, y1, x2, y2 = box[0], box[1], box[0] + box[2], box[1] + box[3]
        return x1, y1, x2, y2, Image.fromarray(impixel[y1:y2, x1:x2])
```

```python
import cv2
cap = cv2.VideoCapture(0)
trainset = load_train_data("./Image")

model = load_model("facenet_keras.h5")
facedetector = MTCNN()
```

```python
while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()
    # Display the resulting frame
    face_rect = extract_face(frame, False)
    if face_rect:
        x1, y1, x2, y2, lolo = face_rect
        frame = cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 0), 1)
        format_face = format_for_facenet(lolo)
        embedding = model.predict(format_face)[0]
        print(EuclideanDistantClassifier(embedding, 12))
        cv2.imshow('frame', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```
